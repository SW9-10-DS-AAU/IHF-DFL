```python
def fed_avg (contrib_scores): 
    n = len(contrib_scores)
    aggregation_scores = [1/n for i in range(n)]
    return aggregation_scores

def positives_only (contrib_scores):
    positive_sum = sum(score for score in contrib_scores if score > 0)
    aggregation_scores = [score / positive_sum if score > 0 else 0 for score in contrib_scores]
    return aggregation_scores

def both (contrib_scores):
    n = len(contrib_scores)
    var = 10000
    aggregation_scores_negatives_normalized = [score if score > 0 else (1 + (1 - abs(score))) / (n * var) for score in contrib_scores]
    # print("aggregation_scores_negative_normalized: ", aggregation_scores_negatives_normalized)
    total_sum = sum(aggregation_scores_negatives_normalized)
    # print("total_sum: ", total_sum)
    
    threshold = 2 / (n * var)
    # print("threshold: ", threshold)

    negative_sum = sum(score for score in aggregation_scores_negatives_normalized if score <= threshold)
    # print("negative_sum: ", negative_sum)
    missing_sum = 1 - negative_sum 
    # print("missing_sum: ", missing_sum)
    positive_sum = total_sum - negative_sum
    
    
    
    aggregation_scores = [((score / positive_sum) * missing_sum) if score > threshold else score for score in aggregation_scores_negatives_normalized]
    
    # print("sum: ", sum(aggregation_scores))
    return aggregation_scores

def plus_one_normalize (contrib_scores):
    n = len(contrib_scores)
    normalized_scores = [score+1 for score in contrib_scores]
    sum_ = n + 1
    aggregation_scores = [score / sum_ for score in normalized_scores]
    return aggregation_scores

def plus_more_than_one_normalize (contrib_scores, more_than_one = 1.1):
    n = len(contrib_scores)
    normalized_scores = [score+more_than_one for score in contrib_scores]
    sum_ = n * more_than_one + 1
    aggregation_scores = [score / sum_ for score in normalized_scores]
    return aggregation_scores
```

```python
contribution_scores = [-1, -0.5, 0, 0.5, 1, 1]
```

```python
print("fed_avg: ", fed_avg(contribution_scores))
print("positives_only: ", positives_only(contribution_scores))
# print("Contribution scores, both: ", both(contribution_scores))
print("plus_one_normalize : ", plus_one_normalize(contribution_scores))
print("plus_more_than_one_normalize : ", plus_more_than_one_normalize(contribution_scores))
```
